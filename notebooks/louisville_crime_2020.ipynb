Parsing the year/month/date/time/day of week from the 2 date fields in Crime Data and moving the new fields to come after their source columns

```python
# Creating new fields function for year, month, day, time, and day of week for both 'Date_Reported' and 'Date_Occurred' in Crime Data for easier analysis of temporal patterns in crime incidents and reordering columns to place these new fields next to the original date fields for easier access and analysis


def date_fields(df):
    
    df = df.copy()

    df['reported_year'] = df['Date_Reported'].dt.year
    df['reported_month'] = df['Date_Reported'].dt.month
    df['reported_day'] = df['Date_Reported'].dt.day
    df['reported_time'] = df['Date_Reported'].dt.time
    df['reported_dow'] = df['Date_Reported'].dt.day_name()

    df['occurred_year'] = df['Date_Occurred'].dt.year
    df['occurred_month'] = df['Date_Occurred'].dt.month
    df['occurred_day'] = df['Date_Occurred'].dt.day
    df['occurred_time'] = df['Date_Occurred'].dt.time
    df['occurred_dow'] = df['Date_Occurred'].dt.day_name()
    
    cols_to_move = ['Date_Reported', 'reported_year', 'reported_month', 'reported_day', 'reported_time', 'reported_dow', 'Date_Occurred', 'occurred_year', 'occurred_month', 'occurred_day', 'occurred_time', 'occurred_dow']

    cols = list(df.columns)

    for col in cols_to_move:
        cols.remove(col)

    insert_at = 3
    for i, col in enumerate(cols_to_move):
        cols.insert(insert_at + i, col)

    df = df[cols]
    
    
# Reordering Crime Data columns to place 'Crime_Type' and 'UOR_Offense_Description' next to other crime classification data for easier analysis

    cols_to_move = ['Crime_Type', 'UOR_Offense_Description']

    cols = list(df.columns)

    for col in cols_to_move:
        cols.remove(col)

    insert_at = 13
    for i, col in enumerate(cols_to_move):
        cols.insert(insert_at + i, col)

    df = df[cols]
    
    return df

split_data_df = date_fields(crime_df)
split_data_df
```

Data cleansing (EDA)

```python
# Cleaning 'ZIP_Code' and 'City' fields function in Crime Data to standardize and prepare for analysis, including handling missing values, correcting known data entry errors, and validating against expected Louisville ZIP Codes (40xxx). This process involves multiple steps such as normalization, imputation, and validation to ensure the integrity of the data for subsequent analysis. The function also includes print statements to show the number of rows and missing ZIP codes before and after cleaning to assess the impact of the cleaning process on the dataset.

def clean_zip_codes(df):
    """
    Cleans and standardizes ZIP_Code field in the crime dataset.
    
    Steps:
    - Normalize format
    - Remove/replace invalid values
    - Impute missing ZIPs using Block_Address and/or LMPD_Beat
    - Validate Louisville ZIPs (40xxx)
    
    Returns cleaned DataFrame.
    """
    
    df = df.copy()

    before = len(df)
    missing_before = df['ZIP_Code'].isna().sum()

    # 1. Normalize format (remove floating point artifacts and convert to string)
    df['ZIP_Code'] = (
        df['ZIP_Code']
        .astype(str)
        .str.replace(r'\.0$', '', regex=True)
    )
    
    # 2. Replace known invalid values with valid values, including replacing 40056 with 40231, which is a valid Louisville ZIP Code, but doesn't represent a specific area of Louisville, and is included in Census Data, and correcting 'DRIVE' and '4291' to valid ZIP Codes based on common data entry errors observed in the dataset.
    df['ZIP_Code'] = (
        df['ZIP_Code']
        .replace('40056', '40231')
        .replace('DRIVE', '40219')
        .replace('4291', '40291')
    )

    # 3. Replace known blank values with NA
    df['ZIP_Code'] = (
        df['ZIP_Code']
        .replace(['00000', '', 'nan', 'NaN', 'None', '<NA>'], pd.NA)
    )
    
    # 4. Replace blank ZIP_Code fields with the most common ZIP_Code for that LMPD_Beat, which may help retain more records in the analysis while acknowledging the limitations of the data. This method may not be accurate for all records, however, as it assumes that the most common ZIP_Code for a given LMPD_Beat is representative of all crimes that occurred in that beat, which may not always be the case.
    df['ZIP_Code'] = (
        df.groupby('LMPD_Beat')['ZIP_Code']
        .transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x)
    )
    
    # 5. Create lookup from most common ZIP for Block_Address
    zip_lookup = (
        df.groupby('Block_Address')['ZIP_Code']
        .agg(lambda x: x.mode().iloc[0] if not x.mode().empty else pd.NA)
    )

    # 6. Fill any additional missing ZIPs using Block_Address
    df['ZIP_Code'] = df.apply(
        lambda row: zip_lookup[row['Block_Address']]
        if pd.isna(row['ZIP_Code']) else row['ZIP_Code'],
        axis=1
    )

    # 7. Standardize City names
    df['City'] = (
        df['City']
        .replace('1719', 'HOLLYVILLA')
        .replace('LOUISV', 'LOUISVILLE')
        .replace('LOUUSVILLE', 'LOUISVILLE')
        .replace('', pd.NA)
        .fillna('LOUISVILLE')
    )

    # 8. Optional fallback (Louisville default ZIP and City)
    df.loc[
        df['Block_Address'].str.upper().str.strip().eq('COMMUNITY AT LARGE')
        & df['ZIP_Code'].isna(),
        'ZIP_Code'
    ] = '40231'

    df.loc[
        (df['City'] == 'LOUISVILLE') & (df['ZIP_Code'].isna()),
        'ZIP_Code'
    ] = '40231'

    # 9. Final validation (keep only Louisville ZIPs)
    df = df[
        df['ZIP_Code'].str.match(r'^40\d{3}$', na=False)
    ]

    after = len(df)
    missing_after = df['ZIP_Code'].isna().sum()
    
    print(f"Rows before: {before}")
    print(f"Rows after: {after}")
    print(f"Missing ZIPs before: {missing_before}")
    print(f"Missing ZIPs after: {missing_after}")
    
    return df

cleaned_df = clean_zip_codes(split_data_df)

cleaned_df

```

```python
# Cleaning data fields function to remove blank values and standardize categorical fields in the crime dataset, including handling missing values in the 'Att_Comp', 'Premise_Type', and 'UCR_Hierarchy' columns by filling them with "Unknown" or "NA" as appropriate, to ensure consistency and integrity of the data for subsequent analysis. This process involves multiple steps such as normalization, imputation, and validation to prepare the dataset for analysis of crime patterns in relation to socioeconomic conditions across different areas of Louisville.

def clean_data_fields(df):
    """
    Cleans and standardizes several fields in the crime dataset to allow for further analysis, including handling missing values and standardizing categorical fields for consistency. This process involves multiple steps such as normalization, imputation, and validation to ensure the integrity of the data for subsequent analysis.
    
    Steps:
    - Fill in blank Attempted/Completed column values with "Unknown"
    - Fill in blank Premise_Type column values with "Unknown"
    - Fill in invalid/blank UCR_Hierarchy column values with "NA"
    - Replace "NA" UCR_Hierarchy column values with "Unknown"
    
    Returns cleaned DataFrame.
    """
    
    df = df.copy()

# 1. Fill in blank Attempted/Completed column values with "Unknown"

    df['Att_Comp'] = (
        df['Att_Comp']
        .replace('', pd.NA)
        .fillna('UNKNOWN')
    )
    
# 2. Fill in blank Premise_Type column values with "Unknown"
    df['Premise_Type'] = (
        df['Premise_Type']
        .replace('', pd.NA)
        .fillna('UNKNOWN')
    )

# 3. Fill in invalid/blank UCR_Hierarchy column values with "NA"
    df['UCR_Hierarchy'] = (
        df['UCR_Hierarchy']
        .replace(['', 'nan', 'NaN', None], pd.NA)
)

# 4. Replace "NA" UCR_Hierarchy column values with "Unknown"
    df['UCR_Hierarchy'] = (
        df['UCR_Hierarchy']
        .fillna('UNKNOWN')
    )

    return df

cleaned2_df = clean_data_fields(cleaned_df)    

cleaned2_df
```

Analysis of Crime Data and Census Data

```python
# Classifying Crime Types: Crime categories were derived using offense descriptions and mapped into violent and property classifications using keyword-based rules aligned with standard criminology definitions. I included all known violent crime types in the keywords list, but this may not be exhaustive and may require further refinement based on the specific crime types present in the dataset and their classifications. Additionally, the classification is based on the presence of keywords in the crime description, which may not always accurately reflect the nature of the crime, so further validation may be necessary to ensure accurate categorization. Much of the "OTHER" category appear to be violent crimes that are not explicitly labeled as such in the dataset, so further analysis may be needed to identify and classify these incidents appropriately.

violent_keywords = [
    "HOMICIDE",
    "ASSAULT",
    "ROBBERY",
    "SEX CRIMES",
    "WEAPONS",
    "ARSON",
    "OTHER"
]

def classify_crime(description):
    if isinstance(description, str):
        desc = description.upper()
        if any(word in desc for word in violent_keywords):
            return "Violent"
    return "Property"

cleaned2_df['Crime_Category'] = cleaned2_df['Crime_Type'].apply(classify_crime)

cleaned2_df
```

```python
# Aggregating Crime Data by ZIP Code to prepare for merging with Census Data and analysis of crime patterns in relation to socioeconomic conditions across different areas of Louisville. This process involves grouping the crime incidents by ZIP Code and calculating relevant metrics such as total crime counts, violent crime counts, property crime counts, and crime rates per capita, which can then be compared against socioeconomic indicators from the Census Data to identify potential correlations and insights.

crime_summary = cleaned2_df.groupby('ZIP_Code').agg(
    Total_Crimes=('Incident_Number', 'count'),
    Violent_Crimes=('Crime_Category', lambda x: (x == 'Violent').sum()),
    Property_Crimes=('Crime_Category', lambda x: (x == 'Property').sum())
).reset_index()

crime_summary
```

```python

# Merging Crime Summary with Census Data to analyze crime patterns in relation to socioeconomic conditions across different areas of Louisville. This process involves combining the aggregated crime data by ZIP Code with the corresponding socioeconomic indicators from the Census Data, such as total population and poverty rate, to enable a comprehensive analysis of how crime rates may correlate with socioeconomic factors in different neighborhoods of Louisville.

merged = crime_summary.merge(
    census_df,
    on='ZIP_Code',
    how='inner'
)

merged['Crime_Rate'] = (merged['Total_Crimes'] / merged['Total_Population']) * 1000
merged['Violent_Rate'] = (merged['Violent_Crimes'] / merged['Total_Population']) * 1000
merged['Property_Rate'] = (merged['Property_Crimes'] / merged['Total_Population']) * 1000

merged = merged.dropna(subset=['Poverty_Rate', 'Crime_Rate'])

merged['Poverty_Group'] = pd.cut(
    merged['Poverty_Rate'],
    bins=[0, 0.1, 0.2, 0.3, 1.0],
    labels=['Low Poverty', 'Moderate', 'High', 'Very High'],
    include_lowest=True
)

merged
```

Plotting Charts

```python
# Chart 1: Scatter plot of Poverty Rate vs Violent Crime Rate to visualize the relationship between poverty and violent crime in Louisville ZIP Codes, with each point representing a ZIP Code and the axes representing the poverty rate and violent crime rate per 1,000 residents, respectively. This visualization can help identify potential correlations or trends between socioeconomic conditions and crime rates across different neighborhoods in Louisville.

plt.figure(figsize=(10,7))
plt.scatter(merged['Poverty_Rate'], merged['Violent_Rate'])
plt.xlabel("Poverty Rate")
plt.ylabel("Violent Crime Rate (per 1,000)")
plt.title("Poverty Rate vs Violent Crime Rate (Louisville, 2020)")
plt.grid(True)
plt.show()
```

```python
# Chart 3: Scatter plot of Poverty Rate vs Property Crime Rate to visualize the relationship between poverty and property crime in Louisville ZIP Codes, with each point representing a ZIP Code and the axes representing the poverty rate and property crime rate per 1,000 residents, respectively. This visualization can help identify potential correlations or trends between socioeconomic conditions and property crime rates across different neighborhoods in Louisville.

plt.figure(figsize=(10,7))
plt.scatter(merged['Poverty_Rate'], merged['Property_Rate'])
plt.xlabel("Poverty Rate")
plt.ylabel("Property Crime Rate")
plt.title("Poverty vs Property Crime Rate (Louisville, 2020)")
plt.show()
```

```python
# Chart 4: Line chart comparing average crime rate and high school graduation rate by ZIP Code to analyze the relationship between education levels and crime rates across different areas of Louisville, with the x-axis representing the ZIP Codes, the primary y-axis representing the average crime rate per 1,000 residents, and a secondary y-axis representing the high school graduation rate as a percentage. This visualization can help identify potential correlations or trends between education levels and crime rates across different neighborhoods in Louisville.

zip_summary = merged.groupby('ZIP_Code').agg({
    'Crime_Rate': 'mean',
    'HS_Grad_Rate': 'mean'
}).reset_index()

corr = zip_summary[['Crime_Rate', 'HS_Grad_Rate']].corr().iloc[0,1]

plt.figure(figsize=(20, 6))

# Primary axis (Crime Rate)
ax1 = plt.gca()
ax1.plot(zip_summary['ZIP_Code'], zip_summary['Crime_Rate'], marker='o', color='blue')
ax1.set_xlabel('ZIP Code', fontweight='bold')
ax1.set_ylabel('Average Crime Rate', fontweight='bold', color='blue')

# Secondary axis (HS Grad Rate)
ax2 = ax1.twinx()
ax2.plot(zip_summary['ZIP_Code'], zip_summary['HS_Grad_Rate'], marker='o', linestyle='--', color='red')
ax2.set_ylabel('HS Graduation Rate', fontweight='bold', color='red')

# Title with correlation coefficient
plt.title(f'Crime Rate vs HS Graduation Rate by ZIP Code (Louisville, 2020)\nCorrelation: {corr:.2f}', fontsize=14)

# Rotate labels for readability
ax1.tick_params(axis='x', rotation=45)

# plt.tight_layout()
plt.show()
```

Database & SQL Coding

```python

import sqlite3
import pandas as pd

conn = sqlite3.connect("../database/louisville_crime.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE locations (
    zip_code TEXT PRIMARY KEY,
    block_address TEXT,
    city TEXT
);
""")

cursor.execute("""
CREATE TABLE crimes (
    crime_id INTEGER PRIMARY KEY AUTOINCREMENT,
    incident_number TEXT,
    zip_code TEXT,
    offense_type TEXT,
    date_reported DATETIME,
    FOREIGN KEY (zip_code) REFERENCES locations(zip_code)
);
""")

cursor.execute("""
CREATE TABLE socioeconomic (
    zip_code TEXT PRIMARY KEY,
    poverty_rate FLOAT,
    hs_grad_rate FLOAT,
    FOREIGN KEY (zip_code) REFERENCES locations(zip_code)
);
""")
```

```python
crime_df = pd.read_csv("../data/crime_2020_-2988378996369536818.csv")
econ_df = pd.read_csv("../data/US_Census_data_Louisville_socio-economic_by_ZIP_Code.csv")

# Normalize column names
crime_df.columns = crime_df.columns.str.strip().str.lower().str.replace(" ", "_")
econ_df.columns = econ_df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
econ_df["poverty_rate"] = (
    econ_df["income_<_poverty_level"] / econ_df["total_estd_population"]
) * 100

econ_df["hs_grad_rate"] = (
    econ_df['inc-<-pl_male_high-school-grad'] + econ_df['inc-<-pl_male_some-college-associates'] + econ_df['inc-<-pl_male_bachelors-higher'] + econ_df['inc-<-pl_female_high-school-grad'] + econ_df['inc-<-pl_female_some-college-associates'] + econ_df['inc-<-pl_female_bachelors-higher'] + econ_df['inc->=-pl_male_high-school-grad'] + econ_df['inc->=-pl_male_some-college-associates'] + econ_df['inc->=-pl_male_bachelors-higher'] + econ_df['inc->=-pl_female_high-school-grad'] + econ_df['inc->=-pl_female_some-college-associates'] + econ_df['inc->=-pl_female_bachelors-higher']
) / econ_df["total_estd_population"] * 100
```
