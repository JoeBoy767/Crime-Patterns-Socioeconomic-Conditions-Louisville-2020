```python
# Creating a DataFrame from the CSV file
df = pd.read_csv('../data/crime_2020_-2988378996369536818.csv')
df.head()
```

```python
# Print column names
df.columns
```

```python
# Renaming Columns for easier access

df.rename(columns={'INCIDENT_NUMBER': 'Incident_Number', 'DATE_REPORTED': 'Date_Reported', 'DATE_OCCURED': 'Date_Occurred', 'CRIME_TYPE': 'Crime_Type', 'UOR_DESC': 'UOR_Offense_Description', 'NIBRS_CODE': 'NIBRS_Code', 'UCR_HIERARCHY': 'UCR_Hierarchy', 'ATT_COMP': 'Att_Comp', 'LMPD_DIVISION': 'LMPD_Division', 'LMPD_BEAT': 'LMPD_Beat', 'PREMISE_TYPE': 'Premise_Type', 'BLOCK_ADDRESS': 'Block_Address', 'ZIP_CODE': 'ZIP_Code'}, inplace=True)
df.columns
```

```python
# Finding and handling missing values
df.isnull().sum()
```

```python
df.isnull().sum() * 100 / len(df)
```

```python
# Identifying data types of each column
df.dtypes
```

```python
# Converting 'Date_Reported' and 'Date_Occurred' to datetime format

df['Date_Reported'] = pd.to_datetime(
    df['Date_Reported'],
    errors='coerce'
)

df['Date_Occurred'] = pd.to_datetime(
    df['Date_Occurred'],
    errors='coerce'
)
df.dtypes
```

Parsing the date/time/day of week from the 2 date fields

```python
df['reported_year'] = df['Date_Reported'].dt.year
df['reported_month'] = df['Date_Reported'].dt.month
df['reported_day'] = df['Date_Reported'].dt.day
df['reported_time'] = df['Date_Reported'].dt.time
df['reported_dow'] = df['Date_Reported'].dt.day_name()
df.head()
```

```python
df['occurred_year'] = df['Date_Occurred'].dt.year
df['occurred_month'] = df['Date_Occurred'].dt.month
df['occurred_day'] = df['Date_Occurred'].dt.day
df['occurred_time'] = df['Date_Occurred'].dt.time
df['occurred_dow'] = df['Date_Occurred'].dt.day_name()
df.head()
```

```python
# Reordering columns to place new date-related columns next to the original date columns

cols_to_move = ['Date_Reported', 'reported_year', 'reported_month', 'reported_day', 'reported_time', 'reported_dow', 'Date_Occurred', 'occurred_year', 'occurred_month', 'occurred_day', 'occurred_time', 'occurred_dow']

cols = list(df.columns)

for col in cols_to_move:
    cols.remove(col)

insert_at = 3
for i, col in enumerate(cols_to_move):
    cols.insert(insert_at + i, col)

df = df[cols]
df.head()
```

```python
cols_to_move = ['Crime_Type', 'UOR_Offense_Description']

cols = list(df.columns)

for col in cols_to_move:
    cols.remove(col)

insert_at = 13
for i, col in enumerate(cols_to_move):
    cols.insert(insert_at + i, col)

df = df[cols]
df.head()
```

```python
# Rediscovering data types after adding new columns
df.dtypes
```

Data cleansing - replacing bad/Null/NaN values

```python
# Cleaning 'ZIP_Code' column - removing floating point artifacts
df['ZIP_Code'] = (
    df['ZIP_Code']
    .astype(str)
    .str.replace(r'\.0$', '', regex=True)
)
df.head()
```

```python
# Cleaning 'ZIP_Code' column - correcting known erroneous entries
df['ZIP_Code'] = (
    df['ZIP_Code']
    .replace('40056', '40201')
    .replace('DRIVE', '40219')
    .replace('4291', '40291')
)

df
```

```python
# Continuing cleaning 'ZIP_Code' column - correcting blank cells, part 1
df['ZIP_Code'] = (
    df['ZIP_Code']
    .replace(['', 'nan', 'NaN', 'None', '<NA>'], pd.NA)
)
df

        
```

```python
# Continuing cleaning 'ZIP_Code' column - correcting blank cells, part 2
df.loc[
    df['Block_Address'].str.upper().str.strip().eq('COMMUNITY AT LARGE')
    & df['ZIP_Code'].isna(),
    'ZIP_Code'
] = '40201'

df
```

```python
#Cleaning dataframe - attempting to find blank ZIP_Code values from other cell values

df['ZIP_Code'] = (
    df.groupby('LMPD_Beat')['ZIP_Code']
      .transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x)
)
df
```

```python
df.loc[
    df['Block_Address'].str.upper().str.strip().eq('INDIAN HILLS TRL / RIVER RD')
    & df['ZIP_Code'].isna(),
    'ZIP_Code'
] = '40207'

df
```

```python
# Cleaning data - correcting known bad City values

df['City'] = (
    df['City']
    .replace('1719', 'HOLLYVILLA')
    .replace('LOUISV', 'LOUISVILLE')
    .replace('LOUUSVILLE', 'LOUISVILLE')
    .replace('', pd.NA)
    .fillna('LOUISVILLE')
)
df
```

```python
# Cleaning data - filling in blank Attempted/Completed column values with "Unknown"

df['Att_Comp'] = (
    df['Att_Comp']
    .replace('', pd.NA)
    .fillna('UNKNOWN')
)
df
```

```python
# Cleaning data - filling in blank Premise_Type column values with "Unknown"

df['Premise_Type'] = (
    df['Premise_Type']
    .replace('', pd.NA)
    .fillna('UNKNOWN')
)
df
```

```python
# Cleaning data - filling in blank UCR_Hierarchy column values with "Unknown", part 1

df['UCR_Hierarchy'] = (
    df['UCR_Hierarchy']
    .replace(['', 'nan', 'NaN', None], pd.NA)
)
df
```

```python
# Cleaning data - filling in blank UCR_Hierarchy column values with "Unknown", part 2

df['UCR_Hierarchy'] = (
    df['UCR_Hierarchy']
    .replace('', pd.NA)
    .fillna('UNKNOWN')
)
df
```

```python
# Get counts of incidents by Crime_Type
crime_counts = df['Crime_Type'].value_counts().reset_index()
crime_counts.columns = ['Crime_Type', 'Count']

# Create bar chart (Demonstrates Comparison)
plt.figure(figsize=(12, 6))
bars = plt.bar(crime_counts['Crime_Type'], crime_counts['Count'], color='skyblue', edgecolor='black')
plt.bar_label(bars, labels=crime_counts['Count'], label_type='edge')  # Add count labels on bars
plt.xticks(rotation=45, ha='right')
plt.xlabel('Crime Type', fontweight="bold")
plt.ylabel('Incident Count', fontweight="bold")
plt.title('Incident Counts by Crime Type, Louisville 2020')
plt.tight_layout()
plt.show()

# In this chart, I wanted to find the answer to the question, "How many incidents occurred per crime type reported in Louisville in 2020?"
```

```python
# Count unique crime types per incident
crime_types_per_incident = df.groupby('Incident_Number')['Crime_Type'].nunique()

# Create histogram (Demonstrates Distribution)
plt.figure(figsize=(8, 5))
counts, bins, patches = plt.hist(crime_types_per_incident, bins=range(1, int(crime_types_per_incident.max()) + 2), edgecolor='black', color='lightcoral', alpha=0.7)
plt.xticks(range(1, int(crime_types_per_incident.max()) + 1))
plt.xlabel('Number of Crime Types per Incident', fontweight="bold")
plt.ylabel('Number of Incidents', fontweight="bold")
plt.title('Crime Types per Incident, Louisville 2020')
plt.tight_layout()

# Add count labels on bars (optional, for clarity)
for i, count in enumerate(counts):
    if count > 0:
        plt.text(bins[i] + 0.5, count + 0.5, str(int(count)), ha='center', va='bottom')

plt.show()

# In this chart, I wanted to find the answer to the question, "How many crime types occurred per incident reported in Louisville in 2020?"

```

```python
# Group by occurred day of week and count
dow_counts = df.groupby('occurred_dow').size().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']).reset_index(name='Incident_Count')

# Create line chart to establish relationship of number of incidents occurred by day of week
plt.figure(figsize=(10, 6))
plt.plot(dow_counts['occurred_dow'], dow_counts['Incident_Count'], marker='o', linestyle='-', color='green')

for x, y in zip(dow_counts['occurred_dow'], dow_counts['Incident_Count']):
    plt.text(x, y, f"{y:,}", ha="center", va="bottom", fontsize=10, fontweight="bold")

plt.xlabel('Day of Week', fontweight="bold")
plt.ylabel('Incident Count', fontweight="bold")
plt.title('Crime Incident Trend by Day of Week (Occurred), Louisville 2020')
plt.grid(True)
plt.ylim(bottom=9000, top=11000)
plt.show()

# In this chart, I wanted to know the answer to, "Does crime occur more frequently during the week (Mon-Fri) vs. the weekend (Sat/Sun)?"
```
